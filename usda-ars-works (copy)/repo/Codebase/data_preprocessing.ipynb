```python
#Data Manipulation
import pandas as pd
import numpy as np
import re
import array
import math
from numpy import NaN
```

```python
df_columns = [
    'Variety',
    'Freq',
    'd(cm)',
    'M%',
    'Density',
    'Attn',
    'Phase',
    'Phase_Corr',
    'Permittivity_real',
    'Permittivity_imaginary'    
]

def is_match(pattern, text):
    match = re.search(pattern, text)
    return bool(match)
```

```python
def check_for_nan(val): #returns true if val is nan
    obj_type = type(val) 

    if obj_type == str or obj_type == np.str_:
        return val == 'nan'
    elif obj_type == float or obj_type == np.float64:
        return math.isnan(val) == True


def process_dataset(df):
    split_data = []
    delimiter = NaN
    for index, row in df.iterrows():
        print('++++++++++++', row)
        grain_name = row['Variety']
        row_arr = np.array(list(row)[1:])

        filtered_row = list(filter(lambda x: not is_match(pattern, str(x)), row_arr))
    
    # trying series
        print('Filtered row: ', filtered_row)
        row_pure = [x for x in filtered_row if not (check_for_nan(x))]

        if row_pure == []: # if full row consists of nan only
            continue

        row_np = np.array(row_pure)
        arr_holy_grail = np.split(row_np, len(row_np)/9)

        for arr in arr_holy_grail:
            demo = list(arr)
            demo.insert(0, grain_name)
            split_data.append(demo)


    new_df = pd.DataFrame(split_data, columns=df_columns)
    return new_df

dataset_file = pd.ExcelFile('usda_dataset.xlsx')
# dataset_file = pd.ExcelFile('sample_dataset.xlsx')
list_grains = dataset_file.sheet_names

for grain_type in list_grains:
    print('===================== Processing GRAIN TYPE: ', grain_type)
    df = dataset_file.parse(grain_type)
    df_processed = process_dataset(df=df)
    df_processed.to_csv('../Datasets/processed/' + grain_type + '.csv')
```
