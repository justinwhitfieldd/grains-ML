```python
import pandas as pd
#Data visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#url dataset
URL = "../Datasets/processed/" + GRAIN_TYPE + ".csv"

#read in excel format
df = pd.read_csv(URL)
df = df[df['Variety'] == 'SOUTH DAKOTA']

print("From USDA: ", URL)

df.head()
```

```python
# Sample data (replace with your data)
columns = [
    'Freq', 
    'Density', 
    'Attn',
    'Phase', 
    'Phase_Corr', 
    'Permittivity_real', 
    'Permittivity_imaginary',
    'M%', 
]

plotTitles = [
    'Frequency', 
    # 'M%', 
    'Bulk Density', 
    'Attenuation',
    'Phase', 
    'Phase Correlation', 
    'Permittivity (real)', 
    'Permittivity (imaginary)',
    'Moisture Content (%)'
]

colors = [
    '#FFB6C1',
    '#AEC6CF',
    '#98FB98',
    '#FFFF99',
    '#B19CD9',
    '#FFD700',
    '#FFDAB9',
    '#ADD8E6'
]

for i, col in enumerate(columns):
    plt.figure(figsize=(5, 6))
    # Create a violin plot
    ax = sns.violinplot(data=df[col], color=colors[i])

    # Remove x-axis ticks
    ax.set_xticks([])

    # Add labels and title
    plt.tight_layout()

    plt.title(plotTitles[i], fontsize=30)
    plt.tick_params(axis='y', labelsize=20)



    # Show the plot
    plt.savefig('saved_figures/violin-plots/' + GRAIN_TYPE + '_' + col + '.svg', 
                bbox_inches='tight',
                dpi=300, 
                # transparent=True
    )
    plt.show()
```
